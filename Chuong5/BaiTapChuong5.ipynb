-----------Bài Làm---------------
"## Một số hàm quan trọng trong xử lý chuỗi của Python\n",
    "\n",
    "- `len(s)`: Trả về độ dài chuỗi `s`.\n",
    "- `s.lower()`: Chuyển toàn bộ chuỗi sang chữ thường.\n",
    "- `s.upper()`: Chuyển toàn bộ chuỗi sang chữ hoa.\n",
    "- `s.capitalize()`: Viết hoa chữ cái đầu tiên của chuỗi.\n",
    "- `s.title()`: Viết hoa chữ cái đầu của mỗi từ trong chuỗi.\n",
    "- `s.strip()`: Loại bỏ khoảng trắng ở đầu và cuối chuỗi.\n",
    "- `s.lstrip()`, `s.rstrip()`: Loại bỏ khoảng trắng bên trái/phải chuỗi.\n",
    "- `s.replace(old, new)`: Thay thế tất cả các chuỗi con `old` bằng `new` trong chuỗi.\n",
    "- `s.split(sep)`: Tách chuỗi thành danh sách các phần tử dựa trên ký tự phân tách `sep`.\n",
    "- `sep.join(list)`: Nối các phần tử trong list thành chuỗi, ngăn cách bởi `sep`.\n",
    "- `s.find(sub)`/`s.index(sub)`: Trả về vị trí xuất hiện đầu tiên của chuỗi con `sub` trong chuỗi, trả về -1 nếu không tìm thấy.\n",
    "\t- `Khác biệt`: `find()` trả về `-1` nếu không tìm thấy, còn `index` sẽ báo lỗi\n",
    "- `s.count(sub)`: Đếm số lần xuất hiện của chuỗi con `sub` trong chuỗi.\n",
    "- `s.startswith(prefix)`, `s.endswith(suffix)`: Kiểm tra chuỗi có bắt đầu/kết thúc bằng chuỗi con chỉ định không.\n",
    "- `s.isdigit()`/`s.isalpha()`/`s.isalnum()`: Kiểm tra chuỗi có phải là số, chữ cái, hay chữ-số.\n"

```python
s=(input("Nhâp vào 1 chuỗi: "))
cntChuHoa =cntChuThuong = cntChuSo = cntKyTu = cntSpace = cntNguyenAm = cntPhuAm = 0
NguyenAm = "aăâáàảãạắằẳẵặấầẩẫậ" \
            "eêéèẻẽẹếềểễệ" \
            "iíìỉĩị" \
            "oôơóòỏõọốồổỗộớờởỡợ" \
            "uưúùủũụứừửữự" \
            "yýỳỷỹỵ" \
            "AĂÂÁÀẢÃẠẮẰẲẴẶẤẦẨẪẬ" \
            "EÊÉÈẺẼẸẾỀỂỄỆ" \
            "IÍÌỈĨỊ" \
            "OÔƠÓÒỎÕỌỐỒỔỖỘỚỜỞỠỢ" \
            "UƯÚÙỦŨỤỨỪỬỮỰ" \
            "YÝỲỶỸỴ"
for x in s:
    if x.isupper():
        cntChuHoa += 1
    elif x.islower():
        cntChuThuong += 1
    elif x.isdigit():
        cntChuSo += 1
    elif x.isspace():
        cntSpace += 1
    elif not x.isalnum():
        cntKyTu += 1
    if x.isalpha():
        if x in NguyenAm:
            cntNguyenAm +=1
        else: cntPhuAm+=1
print("------------------------------------------------------------------------------")
print(f"Chuỗi vừa nhập là: {s}")
print("Số chữ IN HOA:", cntChuHoa)
print("Số chữ in thường:", cntChuThuong)
print("Số chữ là chữ số:", cntChuSo)
print("Số chữ là ký tự đặc biệt:", cntKyTu)
print("Số chữ là khoảng trắng:", cntSpace)
print("Số chữ là nguyên âm (có dấu):", cntNguyenAm)
print("Số chữ là phụ âm:", cntPhuAm)

```

```python
import re

def NegativeNumberInStrings(s):
    # Tìm tất cả các chuỗi có dạng -số, ví dụ: -5, -12
    ket_qua = re.findall(r'-(\d+)', s)

    # Chuyển thành số nguyên âm
    so_nguyen_am = [int("-" + num) for num in ket_qua]
    return so_nguyen_am

s = input("Nhập vào 1 chuỗi: ")
print("Chuỗi vừa nhập là: ",s)
# Xuất kết quả
kq = NegativeNumberInStrings(s)
print("Các số nguyên âm tìm được:", kq)
```

```python
def Toi_Uu_Chuoi(str):
    Chuoi_DanhTu = str.strip().split()
    Chuoi_DanhTu = [x.capitalize() for x in Chuoi_DanhTu]
    return ' '.join(Chuoi_DanhTu)
s = input("Nhập Chuỗi Danh Từ: ")
kq = Toi_Uu_Chuoi(s)
print("Chuỗi vừa nhập là: ",s)
print("Chuỗi tối ưu là: ", kq)
```
